```python
import pandas as pd
```

```python
df_d = pd.read_csv('Annex D.csv', skiprows=4)

#replace NaN with 0
df_d.fillna(0, inplace=True)
df_d.head()
#rename TC_ID to id
df_d.rename(columns={'TC_ID': 'id'}, inplace=True)
```

```python
df_summary = pd.read_csv('dtc_summary.csv')
df = pd.merge(df_d, df_summary, on='id')
df.head()
```

```python
# group by Manouvere and sum the fails and minors. Annoyingly the dataset has different date ranges, hence the division by dailyTestCount. This will mess up DTCs opened half way through the year.
df_global = df.groupby('Manouvere').sum()
df_global['fails'] = df_global['S+D total faults'] / (365 * df_global['dailyTestCount'])
df_global['minors'] = df_global['Driver faults'] / (365 * df_global['dailyTestCount'])
```

```python
df['fails'] = df['S+D total faults'] / (365 * df['dailyTestCount'])
df['minors'] = df['Driver faults'] / (365 * df['dailyTestCount'])
```

```python
#create table. Columns are Manouvere, rows are id, values are fails
df_pivot_f = df.pivot(index=['id','name'], columns='Manouvere', values='fails')
df_pivot_f.fillna(0, inplace=True)
df_pivot_f.to_csv('agg-2023-fails-pivot.csv')

df_pivot_m = df.pivot(index=['id','name'], columns='Manouvere', values='minors')
df_pivot_m.fillna(0, inplace=True)
df_pivot_m.to_csv('agg-2023-minors-pivot.csv')
```
